```python
chess_gray_fl = chess_gray.astype(np.float32)
```

```python
chess_gray_fl
```

```python
res = cv2.cornerHarris(src=chess_gray, blockSize=2, ksize=3, k=0.04)
```

```python
plt.imshow(res)
```

```python
res = cv2.dilate(res,None)
```

```python
chess[res > 0.01*res.max()] = (255,0,0)
```

```python
plt.imshow(chess)
```

```python
chess_real_gray = chess_real_gray.astype(np.float32)
```

```python
resreal = cv2.cornerHarris(src=chess_real_gray, blockSize=3, ksize=3, k=0.04)
```

```python
resreal = cv2.dilate(resreal,None)
```

```python
chess_real[resreal > 0.01*resreal.max()] = (255,0,0)
```

```python
plt.imshow(chess_real)
```

```python
chess = cv2.imread('../DATA/flat_chessboard.png')
chess = cv2.cvtColor(chess, cv2.COLOR_BGR2RGB)

chess_gray = cv2.imread('../DATA/flat_chessboard.png',0)

chess_real = cv2.imread('../DATA/real_chessboard.jpg')
chess_real = cv2.cvtColor(chess_real, cv2.COLOR_BGR2RGB)

chess_real_gray = cv2.cvtColor(chess_real, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(chess_gray,
                                 maxCorners=64, 
                                qualityLevel=0.01, 
                                minDistance=10)
```

```python
corners.shape
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(chess,(x,y),3,(255,0,0), -1)
```

```python
plt.imshow(chess)
```

```python
corners = cv2.goodFeaturesToTrack(chess_real_gray,
                                 maxCorners=300, 
                                qualityLevel=0.01, 
                                minDistance=10)
```

```python
corners.shape
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(chess_real,(x,y),3,(255,0,0), -1)
```

```python
plt.imshow(chess_real)
```
