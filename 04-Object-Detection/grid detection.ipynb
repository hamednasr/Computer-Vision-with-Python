```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']
```

```python
chess = cv2.imread('../DATA/flat_chessboard.png')
chess = cv2.cvtColor(chess, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(chess)
```

```python
found , corners = cv2.findChessboardCorners(chess,(7,7))
```

```python
found
```

```python
corners
```

```python
cv2.drawChessboardCorners(chess,(7,7),corners,found)
```

```python
plt.imshow(chess)
```
