```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']
```

```python
img = cv2.imread('../DATA/sammy_face.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(img,threshold1=240,threshold2=240)
```

```python
plt.imshow(edges)
```

```python
blured_img = cv2.blur(img,ksize=(5,5))
```

```python
plt.imshow(blured_img)
```

```python
edges2 = cv2.Canny(blured_img,threshold1=100,threshold2=150)
```

```python
plt.imshow(edges2)
```
