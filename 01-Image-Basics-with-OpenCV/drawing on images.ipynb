```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']
# from IPython.display import set_matplotlib_formats
# set_matplotlib_formats('svg')
```

```python
blank = np.zeros((512, 512, 3),dtype=np.int16)+255
blank.shape
```

```python
plt.imshow(blank)
```

```python
cv2.rectangle(blank,pt1=(100,300),pt2=(450,450),color=(0,170,120),thickness=5)
```

```python
plt.imshow(blank)
```

```python
cv2.rectangle(blank,pt1=(200,200),pt2=(400,400),color=(0,0,255), thickness=20)
```

```python
plt.imshow(blank)
```

```python
cv2.circle(blank,center=(250,250),radius=100,color=(250,0,0), thickness=10)
```

```python
plt.imshow(blank)
```

```python
cv2.circle(blank,center=(100,100),radius=100,color=(0,255,0), thickness=-1)
```

```python
plt.imshow(blank)
```

```python
cv2.circle(blank,center=(400,100),radius=100,color=(255,255,0), thickness=10)
```

```python
plt.imshow(blank)
```

```python
cv2.line(blank,pt1=(100,500),pt2=(400,0),color=(0,255,0), thickness=10)
```

```python
plt.imshow(blank)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
cv2.putText(blank, text='happy birthday!',org=(0,500),fontFace=font,\
            fontScale=2, color=(255,0,255),thickness=3,lineType=cv2.LINE_AA)
```

```python
plt.imshow(blank)
```
