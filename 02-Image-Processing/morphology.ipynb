```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']
```

```python
def load_img():
    img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img,text='AFGOP',org=(50,300),fontFace=font,fontScale=5,
               color=(255,255,255),thickness=20)
    return img
```

```python
def display_img(img):
    plt.figure(figsize=(12,10))
    plt.imshow(img,cmap='gray')
```

```python
img = load_img()
```

```python
display_img(img)
```

```python
kernel = np.ones((5,5),dtype=np.int8)/20
```

```python
kernel
```

```python
er = cv2.erode(img,kernel,iterations=3 )
```

```python
display_img(er)
```

```python
noise = np.random.randint(0,2,(600,600))*255
noise
```

```python
display_img(noise)
```

```python
img
```

```python
img_whitenoise = img+noise
img_whitenoise
```

```python
img_whitenoise.max()
```

```python
display_img(img_whitenoise)
```

```python
opn = cv2.morphologyEx(img_whitenoise, cv2.MORPH_ELLIPSE,kernel)
```

```python
display_img(opn)
```

```python
bnoise = np.random.randint(0,2,(600,600))*-255
bnoise
```

```python
img_blacknoise = img + bnoise
img_blacknoise
```

```python
display_img(img_blacknoise)
```
